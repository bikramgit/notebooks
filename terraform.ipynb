```python
import os

os.makedirs("terraform_ec2", exist_ok=True)

main_tf = """
provider "aws" {
  region = var.region
}

resource "aws_instance" "vm" {
  ami           = var.ami
  instance_type = var.instance_type
  tags = {
    Name = var.name
  }
}
"""

variables_tf = """
variable "region" {}
variable "ami" {}
variable "instance_type" {}
variable "name" {}
"""

with open("terraform_ec2/main.tf", "w") as f:
    f.write(main_tf)

with open("terraform_ec2/variables.tf", "w") as f:
    f.write(variables_tf)

print("‚úÖ Terraform module created.")
```

```python
import subprocess
from langchain.tools import tool

@tool("terraform_plan_ec2", return_direct=True)
def terraform_plan_ec2(region: str, ami: str, instance_type: str, name: str):
    """Run Terraform plan for an EC2 instance with given parameters."""
    try:
        print("üöÄ Initializing Terraform...")
        subprocess.run(["terraform", "-chdir=terraform_ec2", "init", "-input=false"], check=True)

        print("üßæ Running Terraform plan...")
        plan_cmd = [
            "terraform", "-chdir=terraform_ec2", "plan",
            f"-var=region={region}",
            f"-var=ami={ami}",
            f"-var=instance_type={instance_type}",
            f"-var=name={name}"
        ]
        result = subprocess.run(plan_cmd, capture_output=True, text=True, check=True)

        # Return the plan output
        return result.stdout

    except subprocess.CalledProcessError as e:
        return f"‚ùå Terraform plan failed:\n{e.stderr}"
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.tools import StructuredTool
from langgraph.prebuilt import create_react_agent
import subprocess
import os
import json


def terraform_plan_ec2(region: str, instance_type: str, ami: str, name: str):
    """
    Runs Terraform to plan an EC2 instance deployment in AWS.
    """
    try:
        cmd_init = ["terraform", "-chdir=terraform_ec2", "init", "-input=false"]
        subprocess.run(cmd_init, check=True)

        cmd_plan = [
            "terraform", "-chdir=terraform_ec2", "plan",
            f"-var=region={region}",
            f"-var=instance_type={instance_type}",
            f"-var=ami={ami}",
            f"-var=name={name}"
        ]
        result = subprocess.run(cmd_plan, capture_output=True, text=True, check=True)
        return result.stdout
    except subprocess.CalledProcessError as e:
        return f"Terraform failed: {e.stderr or e.stdout}"



terraform_plan_ec2_tool = StructuredTool.from_function(terraform_plan_ec2)
print("‚úÖ Terraform tool ready.")


```
